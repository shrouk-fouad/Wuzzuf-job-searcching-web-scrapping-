```python
#import modules
import requests
from bs4 import BeautifulSoup
from itertools import zip_longest

import csv
import pandas as pd
```

```python
url = 'https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=business%20analysis'
```

```python
#use requests to get response to reach url
response = requests.get(url)
```

```python
#save page content/markup
src = response.content
```

```python
#check url saved content
src
```

```python
#create soup object to parse content through parser 'lxml'
soup = BeautifulSoup(src,'lxml')
```

```python
#find elemnt contianing job title
job_titles= soup.find_all('h2',{'class':'css-m604qf'})

#check list of job title
job_titles
```

```python
len(job_titles)
```

```python
company_names = soup.find_all('a',{'class':'css-17s97q8'})

company_names
```

```python
location_names = soup.find_all('span',{'class':'css-5wys0k'})


location_names 
```

```python
job_skills = soup.find_all('div',{'class':'css-y4udm8'})


job_skills
```

```python
#create empty lists to fill it with cleaned info
job_title = []
company_name = []
location_name = []
skills = []
links = []
salary = []
requirements = []
```

```python
#loop over returned lists to extract needed text without tags to fill empty lists 
for job in range(len(job_titles)):
    job_title.append(job_titles[job].text)
    company_name.append(company_names[job].text[:-len(' -')])
    location_name.append(location_names[job].text)
    skills.append(job_skills[job].text)
    
    links.append(job_titles[job].find('a').attrs['href'])
```

```python
import re

valid_links = []

for link in links :
    valid_link_format = 'https://wuzzuf.net'+link
    valid_link = re.search('(?P<url>https?://[^\s]+)',valid_link_format).group()
    valid_links.append(valid_link)

valid_links
```

```python
for link in valid_links :
    response = requests.get(link)
    src = response.content
    soup = BeautifulSoup(src,'lxml')
    job_salary = soup.find('sapn',{'class':'css-4xky9y'})
    salary.append(job_salary)
    
#     job_requirements = soup.find('div',{'class':'css-1t5f0fr'})
#     requirements_text = ''
#     for text in job_requirements.find_all('li'):
#         requirements_text+=li.text+'| '
#     requirements.append(requirements_text)
```
